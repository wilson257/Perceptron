# Importing necessary modules

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

# Generating Data

```python
X,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=10,n_redundant=0)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter');
```

# Step Function

```python
def step(z):
    return 1 if z>0 else 0
```

# Perceptron

```python
def perceptron(X,y):
    
    epochs=200
    lr_rate=0.1
    X=np.insert(X,0,1,axis=1)
    weights=np.ones(X.shape[1])
    
    for itr in range(epochs):
        j=np.random.randint(0,100)
        y_hat=step(np.dot(X[j],weights))
        weights=weights+lr_rate*(y[j]-y_hat)*X[j]
        
    return weights[0],weights[1:]
```

```python
intercept_,coef_=perceptron(X,y)
```

```python
#y=mx+c

m=-(coef_[0]/coef_[1])
c=-(intercept_/coef_[1])
```

```python
X_ip=np.linspace(-2,2,200)
y_input=m*X_ip+c
```

# Plotting result

```python
plt.figure(figsize=(10,5))
plt.plot(X_ip,y_input,color='red');
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter');
plt.ylim(-3,2);
```
